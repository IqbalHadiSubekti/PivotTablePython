```python
import pandas as pd
import numpy as np
import requests

chunk = pd.read_csv('hasil.csv',chunksize=1000000, low_memory=False)
df = pd.concat(chunk)
```

```python
df = df.fillna("")
df
```

```python
table = pd.pivot_table(df,index=["Cell Name"],values=["Serving Cell Average RSRP (dBm)","Serving Cell Average RSRQ (dB)"]
                       ,aggfunc=[np.sum,len])
table
```

```python
table.to_csv("ujicoba.csv")
```
